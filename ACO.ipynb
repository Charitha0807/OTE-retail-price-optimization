```python
import numpy as np
import pandas as pd

def run_ant_colony(comp_avg, product_score, iterations=20, ants=10):
    # 1. Define the search space: 50 discrete price points around competitor average
    price_range = np.linspace(comp_avg * 0.7, comp_avg * 1.3, 50)

    # 2. Initialize pheromones (equal at start)
    pheromones = np.ones(len(price_range))
    evaporation_rate = 0.9

    best_price = 0
    max_revenue = 0

    for _ in range(iterations):
        for ant in range(ants):
            # 3. Decision Rule: Select price based on pheromone probability
            probabilities = pheromones / pheromones.sum()
            idx = np.random.choice(len(price_range), p=probabilities)
            current_price = price_range[idx]

            # 4. Fitness Function: Calculate Revenue
            # Logic: Higher score increases demand; higher price decreases it
            base_demand = 100
            price_sensitivity = 2.0
            estimated_qty = base_demand - (price_sensitivity * (current_price - comp_avg)) + (10 * product_score)
            qty = max(0, estimated_qty)
            revenue = current_price * qty

            # 5. Pheromone Update: Deposit based on performance
            pheromones[idx] += (revenue / 100)

            # Track global best
            if revenue > max_revenue:
                max_revenue = revenue
                best_price = current_price

        # 6. Evaporation: Old pheromones fade over time
        pheromones *= evaporation_rate

    return best_price

# --- Example usage with your retail_price.csv data ---
# For a product with competitor average of $90 and a 4.5 star rating:
optimal_price = run_ant_colony(comp_avg=90.0, product_score=4.5)

print(f"The ACO Optimized Price is: ${optimal_price:.2f}")
```
