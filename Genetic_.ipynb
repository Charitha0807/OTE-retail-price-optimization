```python
import numpy as np

def fitness_function(price, comp_avg, product_score):
    # Simplified demand model: Higher price reduces qty, higher score increases it
    base_demand = 100
    demand = base_demand - (2.5 * (price - comp_avg)) + (5 * product_score)
    revenue = max(0, price * demand)
    return revenue

def genetic_algorithm(comp_avg, product_score, generations=50, pop_size=20):
    # Initialize population with random prices around the competitor average
    population = np.random.uniform(comp_avg * 0.8, comp_avg * 1.2, pop_size)

    for _ in range(generations):
        # Calculate fitness
        fitness = np.array([fitness_function(p, comp_avg, product_score) for p in population])

        # Selection (Pick top 50%)
        parents_idx = np.argsort(fitness)[-pop_size//2:]
        parents = population[parents_idx]

        # Crossover & Mutation
        offspring = []
        for i in range(pop_size // 2):
            p1, p2 = np.random.choice(parents, 2)
            child = (p1 + p2) / 2  # Crossover
            child += np.random.normal(0, 2)  # Mutation
            offspring.append(child)

        population = np.concatenate([parents, offspring])

    return population[np.argmax(fitness)]

# Example usage for a specific product
optimal_ga_price = genetic_algorithm(comp_avg=90, product_score=4.5)
print(f"Optimal Price (GA): {optimal_ga_price:.2f}")
```
